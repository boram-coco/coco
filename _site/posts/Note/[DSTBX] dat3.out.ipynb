## 실습34: `gh-pages`에서 만들기 싫다면..?

`-` 생각해보니까 딱히 내 코드가 비밀은 아님

`-` 브랜치를 쓰는게 이론상 깔끔하긴 한데 막상 쓰려니까 불편함. (그
파일이 순간적으로 없어지는 거잖아요?.. 엄청 찝찝해요)

`-` `quarto publish`는 아래동작의 결합이다.

``` bash
quarto render 
# html파일들이 어딘가 만들어져있음. 
git switch gh-pages
# 만들어진 html파일들이 복사됨. 
git add .
git commit -m . 
git push 
git switch main 
```

`-` 아래와 같이 수정하고 싶다면?

``` bash
quarto render 
# html파일들을 docs라는 폴더에 저장
git add .
git commit -m . 
git push 
```

그리고 깃헙설정을 변경한다면?

`-` 하는방법

**step1**

``` bash
git push origin --delete gh-pages # 로칼에서 브랜치삭제 + 원격으로 업데이트
git add .
git commit -m .
git push 
```

**step2**

``` bash
project:
  type: website
  output-dir: docs # <- 이거 추가하세요
# 이 아래는 수정필요없음 
```

**step3**

이후 <https://github.com/%7Buser-name%7D/%7Brepo-name%7D/settings/pages>
로 이동하여 셋팅을 변경하면 된다.
