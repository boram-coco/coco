# 변수선택

## 단계적 전진선택법

```r
m0 = lm(연봉.2018. ~ 1, data = dt)
# 절편먼저 시작
```

```r
add1(m0,
 scope = 연봉.2018. ~팀명+승+패+세+홀드+블론+경기+선발+이닝+삼진.9+볼넷.9+홈런.9+BABIP+LOB.+ERA+RA9.WAR+FIP+kFIP+WAR,
 test = "F") 
```

```r
m1 <- update(m0, ~ . +WAR)
summary(m1)
```

```r
add1(m1,
 scope = 연봉.2018. ~팀명+승+패+세+홀드+블론+경기+선발+이닝+삼진.9+볼넷.9+홈런.9+BABIP+LOB.+ERA+RA9.WAR+FIP+kFIP+WAR,
 test = "F") 
```

- 너무 귀찮고... 복잡한데..

## AIC이용

### STEP

```r
model_step = step(
 m0,
 scope = 연봉.2018. ~팀명+승+패+세+홀드+블론+경기+선발+이닝+삼진.9+볼넷.9+홈런.9+BABIP+LOB.+ERA+RA9.WAR+FIP+kFIP+WAR,
 direction = "both")
summary(model_step)
```

- AIC를 이용하면 최종 모형은 "연봉.2018. ~ WAR + 세 + 승 + 패 + RA9.WAR" 이다.

### 후진

```r
model_back = step(model2, direction = "backward")
summary(model_back)
```

### forwrad

```r
model_forward = step(
 m0,
 scope = 연봉.2018. ~팀명+승+패+세+홀드+블론+경기+선발+이닝+삼진.9+볼넷.9+홈런.9+BABIP+LOB.+ERA+RA9.WAR+FIP+kFIP+WAR,
 direction = "forward")
summary(model_forward)
```
