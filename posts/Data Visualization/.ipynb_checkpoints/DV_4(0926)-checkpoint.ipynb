```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# 산점도 응용예제4 (무상관과 독립)

## 예시자료

`-` 예시1

```python
x1 = np.random.uniform(low=0, high=1, size=10000)  # 0~1사이의 10000개
y1 = np.random.uniform(low=0, high=1, size=10000) 
```

```python
plt.plot(x1,y1,',')
plt.plot(x1[0], y1[0],'or')
```

`-` 예시2 (원)

```python
_r2 = x1**2 + y1**2 # 반지름
_r2
```

```python
len(_r2)
```

```python
x2=x1[_r2<1]
y2=y1[_r2<1]
```

```python
plt.plot(x2,y2,',')
```

`-` 예시3 (이변량정규분포)

```python
x3 = np.random.randn(10000)
y3 = np.random.randn(10000)
```

```python
plt.plot(x3,y3,',')
```

## 상관계수

예시1, 예시2, 예시3의 산점도를 보고 상관계수가 얼마인지 예상해보라. 실제 계산결과와 확인하라.

```python
np.corrcoef([x1,y1])
```

```python
np.corrcoef([x2,y2])
```

```python
np.corrcoef([x3,y3])
```

독립을 따져보자.

```python
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].plot(x1,y1,',')
ax[1].plot(x2,y2,',')
```

```python
def g(intval, data, ax, col = 'r'):
    a,b = intval
    x,y = data
    idx = (a<x) & (x<b)
    ax.plot(x[idx], y[idx],',',color=col)
```

```python
fig
```

```python
g([-0.1,0.1],[x1,y1],ax[0])
g([-0.1,0.1],[x2,y2],ax[1])
fig
```

```python
g([0.79,0.99],[x1,y1],ax[0],col='m')
g([0.79,0.99],[x2,y2],ax[1],col='m')
fig
```
