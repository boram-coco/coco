```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

# Seaborn 특징

`-` 특1: 입력으로 데이터프레임을 선호한다. (matplotlib은 array를 선호)

- 그렇다고 해서 데이터프레임이 아닌 경우 그림이 아예 안 그려지지는 않는다.
- 데이터프레임 형태는 long form 과 wide form 이 있다. (ref:https://seaborn.pydata.org/tutorial/data_structure.html#long-form-vs-wide-form-data)
- 참고로 long form이 더 우수한 저장형태!
- wide-df = [array1,array2,array3]
- long-df = [array_val, array_cat]

`-` 특2: matplotlib을 존경함. (ref:https://seaborn.pydata.org/)

# sns boxplot

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
y1=[75,75,76,76,77,77,79,79,79,98] # A선생님에게 통계학을 배운 학생의 점수들
y2=[76,76,77,77,78,78,80,80,80,81] # B선생님에게 통계학을 배운 학생의 점수들 
```

## plt복습

```python
plt.boxplot([y1,y2])
```

## sns: wide df

```python
df1=pd.DataFrame({1:y1,2:y2})
df1
```

`-` 예시1

```python
sns.boxplot(data=df1)
```

```python
sns.boxplot(np.stack([y1,y2],axis=1))   # 잘 쓰진 않는데 된다
```

## sns: long df

value 를 넣고 그 value가 어떤 category에 있는지 넣는 방법

```python
df2 = pd.DataFrame({'score': y1+y2, 'class': ['A']*len(y1)+['B']*len(y2)})
df2
```

`-` 예시1

```python
sns.boxplot(data=df2, x='class', y='score')
```

## sns: array

`-` 예시1

```python
sns.boxplot(data=y1)
```

`-` 예시2

```python
sns.boxplot(y=y1)
```

`-` 예시3

```python
sns.boxplot(x=y1)
```

# sns histplot

`-` 데이터

```python
x = np.random.randn(10000)
y = np.random.randn(10000) + 1
```

## plt 복습

```python
plt.hist(x,alpha=0.5)
plt.hist(y,alpha=0.5)
```

`-` 예시2

```python
plt.hist([x,y]);
```

## sns: wide df

```python
df1=pd.DataFrame({'x':x, 'y':y})
df1
```

`-` 예시

```python
sns.histplot(data=df1);
```

```python
sns.histplot(data=df1,bins=20)  # 칸 조정
```

```python
sns.histplot(data=df1,bins=20,kde=True);    # kde : 곡선
```

```python
sns.histplot(data=df1,bins=20,kde=True,element="step");
```

```python
sns.histplot(data=df1,bins=20,kde=True,element="step",lw=5) # mpl에 대한 존경심 확인 
```

## sns: long df

```python
df2=pd.DataFrame({'val' : np.concatenate([x,y]), 'var': ['x']*len(x) + ['y']*len(y)})
df2
```

```python
sns.histplot(data=df2, x='val', hue='var', bins=20, kde=True, lw=0)
# hue:색깔 var로 구분하겠다! 
```

## sns: array

```python
sns.histplot(data=x)
```

```python
sns.histplot(x=x)
```

```python
sns.histplot(x=x, color='C0', bins=20, lw=0)
sns.histplot(x=y, color='C0', bins=20, lw=0)
```

# sns lineplot

```python
np.random.seed(43052)
ϵ = np.random.randn(100)
```

```python
y = np.cumsum(ϵ)  # 누적값 더하기
```

## plt복습

```python
plt.plot(ϵ,'--o')
```

```python
plt.plot(y,'--o')
```

## sns: array

```python
sns.lineplot(data=ϵ)
```

```python
sns.lineplot(data=y)
```

## sns: wide df

```python
df4 = pd.DataFrame({'ϵ' : ϵ , 'y' : y})
df4
```

```python
sns.lineplot(data=df4)
```

```python
sns.lineplot(data=df4, dashes=False)   # dashes : 둘다 실선
```

## sns: long df

```python
df5= pd.DataFrame({'idx':list(range(100))*2,'val':np.concatenate([ϵ,y]),'cat':['eps']*100 + ['y']*100 })
df5
```

```python
sns.lineplot(data=df5, x='idx',y='val',hue='cat')
```

```python
sns.lineplot(data=df5, x='idx',y='val',style='cat',hue='cat',markers=True)
```

```python
sns.lineplot(data=df5, x='idx',y='val',style='cat',hue='cat',dashes=[(3,1),(3,3)],markers=['o','o'])
```
