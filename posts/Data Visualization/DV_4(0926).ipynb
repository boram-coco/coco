# 산점도 응용예제4 (무상관과 독립)

## 예시자료

`-` 예시1

```python
x1 = np.random.uniform(low=0, high=1, size=10000)  # 0~1사이의 10000개
y1 = np.random.uniform(low=0, high=1, size=10000) 
```

```python
plt.plot(x1,y1,',')
plt.plot(x1[0], y1[0],'or')
```

`-` 예시2 (원)

```python
_r2 = x1**2 + y1**2 # 반지름
_r2
```

```python
len(_r2)
```

```python
x2=x1[_r2<1]
y2=y1[_r2<1]
```

```python
plt.plot(x2,y2,',')
```

`-` 예시3 (이변량정규분포)

```python
x3 = np.random.randn(10000)
y3 = np.random.randn(10000)
```

```python
plt.plot(x3,y3,',')
```

## 상관계수

예시1, 예시2, 예시3의 산점도를 보고 상관계수가 얼마인지 예상해보라. 실제 계산결과와 확인하라.

```python
np.corrcoef([x1,y1])
```

```python
np.corrcoef([x2,y2])
```

```python
np.corrcoef([x3,y3])
```

독립을 따져보자.

```python
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].plot(x1,y1,',')
ax[1].plot(x2,y2,',')
```

```python
def g(intval, data, ax, col = 'r'):
    a,b = intval
    x,y = data
    idx = (a<x) & (x<b)
    ax.plot(x[idx], y[idx],',',color=col)
```

```python
fig
```

```python
g([-0.1,0.1],[x1,y1],ax[0])
g([-0.1,0.1],[x2,y2],ax[1])
fig
```

```python
g([0.79,0.99],[x1,y1],ax[0],col='m')
g([0.79,0.99],[x2,y2],ax[1],col='m')
fig
```

`-` 예시3

```python
plt.plot(x3,y3,',')
```

```python
fig, ax = plt.subplots()
ax.plot(x3,y3,',',color='gray')     # g함수 사용하기 위해 이렇게 해보자
```

```python
g([-2.5,-1.5],[x3,y3],ax,col='r')
fig
```

```python
g([-2.5,-1.5],[x3,y3],ax,col='r')
g([-0.5,+0.5],[x3,y3],ax,col='b')
g([+1.5,+2.5],[x3,y3],ax,col='g')
fig
```

```python
idx = (-0.5<x3) & (x3<0.5)
plt.hist(y3[idx])  # 위의 파란색을 뽑아보자
```

```python
idx = (-0.5<x3) & (x3<0.5)
plt.hist(y3[idx]);  # 맨 뒤 세미콜론 붙이면 위의 글씨 안나옴
```

```python
def h(intval, data, ax, col):
    a,b = intval
    x,y = data
    idx = (a<x) &  (x<b)
    ax.hist(y[idx], color=col)
```

```python
fig,ax = plt.subplots(5,2,figsize=(8,16))
ax[0,0].plot(x3,y3,',',color='gray'); g([-2.5,-1.5],[x3,y3],ax[0,0],col='r')
ax[1,0].plot(x3,y3,',',color='gray'); g([-1.5,-0.5],[x3,y3],ax[1,0],col='g')
ax[2,0].plot(x3,y3,',',color='gray'); g([-0.5,+0.5],[x3,y3],ax[2,0],col='b')
ax[3,0].plot(x3,y3,',',color='gray'); g([+0.5,+1.5],[x3,y3],ax[3,0],col='m')
ax[4,0].plot(x3,y3,',',color='gray'); g([+1.5,+2.5],[x3,y3],ax[4,0],col='lime')

h([-2.5,-1.5],[x3,y3],ax[0,1],col='r')
h([-1.5,-0.5],[x3,y3],ax[1,1],col='g')
h([-0.5,+0.5],[x3,y3],ax[2,1],col='b')
h([+0.5,+1.5],[x3,y3],ax[3,1],col='m')
h([+1.5,+2.5],[x3,y3],ax[4,1],col='lime')
```
